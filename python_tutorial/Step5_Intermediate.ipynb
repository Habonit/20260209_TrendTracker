````python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ğŸ“– ì´ë¡ : raiseì™€ ì‹¤ì „ ì˜ˆì‹œ
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

# raise: ì˜ë„ì ìœ¼ë¡œ ì—ëŸ¬ ë°œìƒ
def set_age(age):
    if age < 0:
        raise ValueError("ë‚˜ì´ëŠ” ìŒìˆ˜ê°€ ë  ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
    return age

try:
    set_age(-5)
except ValueError as e:
    print(f"ì—ëŸ¬: {e}")

# ì‹¤ì „: ì•ˆì „í•œ ë‚˜ëˆ„ê¸°
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return None

print(f"10 / 3 = {safe_divide(10, 3):.2f}")
print(f"10 / 0 = {safe_divide(10, 0)}")
````
