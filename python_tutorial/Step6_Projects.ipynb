```python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ğŸ“– ìƒ˜í”Œ ë°ì´í„° ìƒì„±
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

import pandas as pd
import random

# ìƒ˜í”Œ íŒë§¤ ë°ì´í„° ìƒì„±
random.seed(42)

products = ["ë…¸íŠ¸ë¶", "ë§ˆìš°ìŠ¤", "í‚¤ë³´ë“œ", "ëª¨ë‹ˆí„°", "í—¤ë“œì…‹"]
regions = ["ì„œìš¸", "ë¶€ì‚°", "ëŒ€êµ¬", "ì¸ì²œ"]

data = {
    "ë‚ ì§œ": [],
    "ì œí’ˆ": [],
    "ì§€ì—­": [],
    "ìˆ˜ëŸ‰": [],
    "ë‹¨ê°€": [],
}

prices = {"ë…¸íŠ¸ë¶": 1200000, "ë§ˆìš°ìŠ¤": 35000, "í‚¤ë³´ë“œ": 89000, "ëª¨ë‹ˆí„°": 450000, "í—¤ë“œì…‹": 120000}

for month in range(1, 13):
    for _ in range(8):
        product = random.choice(products)
        data["ë‚ ì§œ"].append(f"2025-{month:02d}-{random.randint(1,28):02d}")
        data["ì œí’ˆ"].append(product)
        data["ì§€ì—­"].append(random.choice(regions))
        data["ìˆ˜ëŸ‰"].append(random.randint(1, 10))
        data["ë‹¨ê°€"].append(prices[product])

df = pd.DataFrame(data)
df["ë§¤ì¶œ"] = df["ìˆ˜ëŸ‰"] * df["ë‹¨ê°€"]
df.to_csv("sales_data.csv", index=False, encoding="utf-8-sig")
print(f"âœ… ìƒ˜í”Œ ë°ì´í„° ìƒì„± ì™„ë£Œ! ({len(df)}í–‰)")
print(df.head())
```

```python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ğŸ“– ì´ë¡ : ì‹œê°í™” (matplotlib)
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams["font.family"] = "DejaVu Sans"

# ì œí’ˆë³„ ë§¤ì¶œ ë§‰ëŒ€ê·¸ë˜í”„
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1. ì œí’ˆë³„ ì´ ë§¤ì¶œ
product_sales = df.groupby("ì œí’ˆ")["ë§¤ì¶œ"].sum().sort_values(ascending=True)
product_sales.plot(kind="barh", ax=axes[0], color="steelblue")
axes[0].set_title("Product Sales (Total)")
axes[0].set_xlabel("Sales (KRW)")

# 2. ì§€ì—­ë³„ íŒë§¤ ê±´ìˆ˜
region_counts = df["ì§€ì—­"].value_counts()
region_counts.plot(kind="bar", ax=axes[1], color="coral")
axes[1].set_title("Sales Count by Region")
axes[1].set_ylabel("Count")
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```
