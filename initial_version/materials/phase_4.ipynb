## 📚 파일 입출력이란?

### 정의

파일 입출력은 **프로그램에서 파일에 데이터를 저장하거나 읽어오는 작업**입니다.

| 함수 | 역할 |
|------|------|
| `open()` | 파일 열기 |
| `write()` | 파일에 쓰기 |
| `read()` | 파일 읽기 |
| `close()` | 파일 닫기 |

프로그램이 종료되면 메모리의 데이터는 사라지지만,
**파일에 저장하면 영구적으로 보관**할 수 있습니다.

### 왜 필요한가요?

1. **데이터 보존**: 프로그램을 껐다 켜도 데이터 유지
2. **데이터 공유**: 다른 프로그램과 데이터 교환
3. **백업**: 중요한 정보를 안전하게 저장

### 실행해서 확인해봅시다!

````python
# ============================================
# 🔍 파일 쓰기 (Write) 기본
# ============================================
# 텍스트 파일에 데이터를 저장해봅시다.

# 방법 1: 기본 방식 (close 필수)
file = open("test_basic.txt", "w", encoding="utf-8")  # w = write 모드
file.write("안녕하세요!\n")  # 파일에 쓰기
file.write("파이썬 파일 입출력입니다.\n")
file.close()  # 반드시 닫아야 함!

print("✅ test_basic.txt 파일이 생성되었습니다.")

# 방법 2: with 문 사용 (권장! - 자동으로 닫힘)
with open("test_with.txt", "w", encoding="utf-8") as file:
    file.write("with 문을 사용하면\n")
    file.write("자동으로 파일이 닫힙니다!\n")
# 여기서 자동으로 close() 됨!

print("✅ test_with.txt 파일이 생성되었습니다.")

# 💡 with 문을 사용하면 close()를 잊어도 안전합니다!
````

````python
# ============================================
# 🔍 파일 읽기 (Read) 기본
# ============================================
# 저장한 파일을 읽어봅시다.

# 전체 읽기
print("=== 전체 읽기 ===")
with open("test_with.txt", "r", encoding="utf-8") as file:  # r = read 모드
    content = file.read()  # 전체 내용을 문자열로
    print(content)

# 한 줄씩 읽기
print("=== 한 줄씩 읽기 ===")
with open("test_with.txt", "r", encoding="utf-8") as file:
    for i, line in enumerate(file, 1):  # 파일을 한 줄씩 반복
        print(f"  {i}번째 줄: {line.strip()}")  # strip()으로 줄바꿈 제거

# 💡 파일 모드 정리:
# - "r": 읽기 (파일 없으면 에러)
# - "w": 쓰기 (파일 있으면 덮어씀)
# - "a": 추가 (파일 끝에 이어씀)
# - "r+": 읽기+쓰기
````

````python
# ============================================
# 🔍 파일/폴더 존재 확인 및 생성
# ============================================
# os 모듈로 파일 시스템을 다룹니다.

import os

# 파일 존재 확인
print("=== 파일 존재 확인 ===")
print(f"test_with.txt 존재: {os.path.exists('test_with.txt')}")  # True
print(f"없는파일.txt 존재: {os.path.exists('없는파일.txt')}")    # False

# 폴더 생성
print("\n=== 폴더 생성 ===")
os.makedirs("test_folder/sub_folder", exist_ok=True)  # 중첩 폴더도 한번에!
# exist_ok=True: 이미 있어도 에러 안 남
print(f"폴더 생성됨: {os.path.exists('test_folder/sub_folder')}")

# 경로에서 폴더 부분 추출
print("\n=== 경로 분리 ===")
path = "data/search_history.csv"
folder = os.path.dirname(path)  # "data"
filename = os.path.basename(path)  # "search_history.csv"
print(f"전체 경로: {path}")
print(f"폴더: {folder}")
print(f"파일명: {filename}")

# 💡 os.makedirs()로 폴더를 자동 생성하면
# 파일 저장 전에 폴더가 없어서 에러나는 것을 방지합니다!
````

### 📝 핵심 정리

| 함수 | 역할 |
|------|------|
| `open(path, mode)` | 파일 열기 |
| `file.write(text)` | 파일에 쓰기 |
| `file.read()` | 파일 전체 읽기 |
| `with open() as f:` | 안전한 파일 사용 (자동 close) |
| `os.path.exists()` | 파일/폴더 존재 확인 |
| `os.makedirs()` | 폴더 생성 |

💡 **기억하세요**: `with` 문을 사용하면 파일을 안전하게 다룰 수 있습니다!

## 📚 CSV 파일이란?

### 정의

CSV는 **쉼표로 구분된 값들을 저장하는 텍스트 파일 형식**입니다.

- **C**omma **S**eparated **V**alues (쉼표로 구분된 값들)
- 텍스트 파일이지만 표 형태의 데이터를 저장
- 엑셀, 구글 시트 등에서 열 수 있음

```
이름,나이,도시
홍길동,25,서울
김철수,30,부산
```

### 왜 필요한가요?

1. **단순함**: 메모장으로도 열어볼 수 있음
2. **호환성**: 거의 모든 프로그램에서 지원
3. **가벼움**: 복잡한 형식 없이 데이터만 저장

### 실행해서 확인해봅시다!

````python
# ============================================
# 🔍 CSV 파일 직접 만들어보기
# ============================================
# 순수 파이썬으로 CSV 파일을 만들어봅시다.

# CSV 내용 작성
csv_content = """이름,나이,도시
홍길동,25,서울
김철수,30,부산
이영희,28,인천"""

# 파일로 저장
with open("people.csv", "w", encoding="utf-8-sig") as file:
    # utf-8-sig: 엑셀에서 한글 깨짐 방지
    file.write(csv_content)

print("✅ people.csv 파일이 생성되었습니다.")
print("\n=== 파일 내용 ===")
print(csv_content)

# 💡 쉼표(,)로 값을 구분하고, 줄바꿈으로 행을 구분합니다!
````

````python
# ============================================
# 🔍 CSV 모듈로 읽고 쓰기
# ============================================
# 파이썬 내장 csv 모듈을 사용해봅시다.

import csv

# CSV 읽기
print("=== CSV 읽기 (csv 모듈) ===")
with open("people.csv", "r", encoding="utf-8-sig") as file:
    reader = csv.reader(file)  # CSV 리더 생성
    for i, row in enumerate(reader):
        if i == 0:
            print(f"헤더: {row}")
        else:
            print(f"데이터 {i}: {row}")

# 딕셔너리 형태로 읽기 (더 편리!)
print("\n=== CSV 읽기 (DictReader) ===")
with open("people.csv", "r", encoding="utf-8-sig") as file:
    reader = csv.DictReader(file)  # 헤더를 키로 사용
    for row in reader:
        print(f"{row['이름']}님은 {row['나이']}세, {row['도시']} 거주")

# 💡 DictReader를 사용하면 컬럼 이름으로 값에 접근할 수 있습니다!
````

### 📝 핵심 정리

| 항목 | 설명 |
|------|------|
| **CSV** | 쉼표로 구분된 텍스트 파일 (표 형태) |
| **헤더** | 첫 번째 줄, 컬럼 이름들 |
| **csv.reader** | CSV를 리스트로 읽기 |
| **csv.DictReader** | CSV를 딕셔너리로 읽기 |
| **utf-8-sig** | 엑셀 호환 인코딩 |

💡 **기억하세요**: CSV는 가장 간단한 표 형식 데이터 저장 방법입니다!

## 📚 pandas DataFrame이란?

### 정의

pandas DataFrame은 **행과 열로 구성된 2차원 표 형태의 데이터 구조**입니다.

| pandas | 설명 |
|--------|------|
| DataFrame | 2차원 표 데이터 |
| Index | 행 식별자 |
| Column | 열 이름 |
| Value | 셀 값 |

### 왜 필요한가요?

1. **CSV 쉽게 다루기**: 한 줄로 읽고 쓰기
2. **강력한 기능**: 필터링, 정렬, 집계 등
3. **데이터 분석**: 통계, 시각화 연동

### 실행해서 확인해봅시다!

````python
# ============================================
# 🔍 pandas DataFrame 기본
# ============================================
# DataFrame을 만들고 다루는 방법을 알아봅시다.

import pandas as pd  # pandas는 보통 pd로 줄여서 사용

# 방법 1: 딕셔너리로 만들기
data = {
    "이름": ["홍길동", "김철수", "이영희"],
    "나이": [25, 30, 28],
    "도시": ["서울", "부산", "인천"]
}
df = pd.DataFrame(data)

print("=== DataFrame 생성 ===")
print(df)
print(f"\n크기: {df.shape}")  # (행 수, 열 수)
print(f"컬럼: {list(df.columns)}")
````

````python
# ============================================
# 🔍 DataFrame 데이터 접근
# ============================================
# DataFrame에서 원하는 데이터를 꺼내봅시다.

import pandas as pd

df = pd.DataFrame({
    "이름": ["홍길동", "김철수", "이영희"],
    "나이": [25, 30, 28],
    "도시": ["서울", "부산", "인천"]
})

# 컬럼 접근 (Series 반환)
print("=== 컬럼 접근 ===")
print(df["이름"])  # 이름 컬럼만

# 행 접근 (iloc: 숫자 인덱스, loc: 레이블)
print("\n=== 행 접근 ===")
print(f"첫 번째 행: {df.iloc[0].to_dict()}")
print(f"두 번째 행: {df.iloc[1].to_dict()}")

# 특정 값 접근
print("\n=== 특정 값 접근 ===")
print(f"1행 2열(나이): {df.iloc[0]['나이']}")
print(f"김철수의 도시: {df[df['이름'] == '김철수']['도시'].values[0]}")
````

````python
# ============================================
# 🔍 DataFrame 필터링과 정렬
# ============================================
# 조건에 맞는 데이터를 찾고 정렬해봅시다.

import pandas as pd

df = pd.DataFrame({
    "이름": ["홍길동", "김철수", "이영희", "박민수"],
    "나이": [25, 30, 28, 22],
    "도시": ["서울", "부산", "서울", "대전"]
})

# 필터링: 조건에 맞는 행만
print("=== 필터링 ===")
print("서울 거주자:")
print(df[df["도시"] == "서울"])

print("\n25세 이상:")
print(df[df["나이"] >= 25])

# 정렬
print("\n=== 정렬 ===")
print("나이 오름차순:")
print(df.sort_values("나이"))

print("\n나이 내림차순:")
print(df.sort_values("나이", ascending=False))
````

````python
# ============================================
# 🔍 pandas로 CSV 읽고 쓰기
# ============================================
# pandas를 사용하면 CSV를 아주 쉽게 다룰 수 있습니다!

import pandas as pd

# CSV 읽기 (한 줄!)
print("=== CSV 읽기 ===")
df = pd.read_csv("people.csv", encoding="utf-8-sig")
print(df)

# 데이터 추가
new_row = pd.DataFrame([{"이름": "박지민", "나이": 27, "도시": "대구"}])
df = pd.concat([df, new_row], ignore_index=True)
print("\n=== 행 추가 후 ===")
print(df)

# CSV 쓰기 (한 줄!)
df.to_csv("people_updated.csv", index=False, encoding="utf-8-sig")
print("\n✅ people_updated.csv 저장 완료!")

# 💡 pandas를 사용하면 CSV 한 줄로 읽고 쓸 수 있습니다!
````

## 📚 Repository 패턴이란?

### 정의

Repository 패턴은 **데이터 저장소 접근 로직을 캡슐화하여 일관된 인터페이스를 제공하는 디자인 패턴**입니다.

| 메서드 | 역할 |
|--------|------|
| `load()` | 데이터 불러오기 |
| `save()` | 데이터 저장 |
| `find_by_key()` | 키로 찾기 |
| `get_all_keys()` | 전체 키 목록 조회 |

**핵심**: 데이터 저장 방식(CSV, DB 등)을 숨기고, 간단한 인터페이스만 제공!

### 왜 필요한가요?

```python
# Repository 없이 (복잡!)
df = pd.read_csv("data.csv")
filtered = df[df["key"] == "abc"]
# ... 복잡한 변환 로직 ...

# Repository 사용 (간단!)
result = repository.find_by_key("abc")
```

1. **추상화**: 저장 방식을 몰라도 됨
2. **일관성**: 데이터 접근 방법이 통일됨
3. **변경 용이**: CSV → DB 변경해도 사용하는 코드는 그대로

### 실행해서 확인해봅시다!

````python
# ============================================
# 🔍 Repository 패턴 이해하기
# ============================================
# 간단한 Repository 예시를 만들어봅시다.

import os
import pandas as pd
from dataclasses import dataclass
from typing import List, Optional

# 도메인 모델
@dataclass
class Product:
    id: str
    name: str
    price: int

# Repository 클래스
class ProductRepository:
    """상품 데이터를 관리하는 창고 관리자"""
    
    def __init__(self, csv_path: str):
        self.csv_path = csv_path
        # 폴더가 없으면 생성
        folder = os.path.dirname(csv_path)
        if folder:
            os.makedirs(folder, exist_ok=True)
    
    def load(self) -> pd.DataFrame:
        """데이터 로드"""
        if not os.path.exists(self.csv_path):
            return pd.DataFrame(columns=["id", "name", "price"])
        return pd.read_csv(self.csv_path)
    
    def save(self, product: Product) -> bool:
        """상품 저장"""
        df = self.load()
        new_row = pd.DataFrame([{"id": product.id, "name": product.name, "price": product.price}])
        df = pd.concat([df, new_row], ignore_index=True)
        df.to_csv(self.csv_path, index=False)
        return True
    
    def find_by_id(self, product_id: str) -> Optional[Product]:
        """ID로 상품 찾기"""
        df = self.load()
        filtered = df[df["id"] == product_id]
        if filtered.empty:
            return None
        row = filtered.iloc[0]
        return Product(id=row["id"], name=row["name"], price=int(row["price"]))
    
    def get_all(self) -> List[Product]:
        """모든 상품 조회"""
        df = self.load()
        return [Product(id=row["id"], name=row["name"], price=int(row["price"])) 
                for _, row in df.iterrows()]

print("✅ ProductRepository 클래스가 정의되었습니다!")
````

````python
# ============================================
# 🔍 Repository 사용해보기
# ============================================
# 위에서 만든 Repository를 사용해봅시다.

# Repository 생성
repo = ProductRepository("test_data/products.csv")

# 상품 저장
print("=== 상품 저장 ===")
repo.save(Product("P001", "노트북", 1500000))
repo.save(Product("P002", "마우스", 50000))
repo.save(Product("P003", "키보드", 100000))
print("✅ 3개 상품 저장 완료!")

# 전체 조회
print("\n=== 전체 상품 조회 ===")
products = repo.get_all()
for p in products:
    print(f"  - {p.name}: {p.price:,}원")

# ID로 검색
print("\n=== ID로 검색 ===")
found = repo.find_by_id("P002")
if found:
    print(f"찾음: {found.name} ({found.price:,}원)")

not_found = repo.find_by_id("P999")
print(f"P999 검색 결과: {not_found}")

# 💡 Repository를 사용하면 데이터 저장/조회가 간단해집니다!
# CSV가 DB로 바뀌어도 사용하는 코드는 그대로!
````

### 📝 핵심 정리

| 메서드 | 역할 |
|--------|------|
| `load()` | 데이터 불러오기 |
| `save()` | 데이터 저장 |

## 🔎 repositories/search_repository.py 분석

**파일**: `repositories/search_repository.py`
**역할**: 검색 결과를 CSV 파일에 저장하고 조회하는 Repository

````python
# ============================================
# 🔎 initial_version의 repositories/search_repository.py 코드
# ============================================
# 이 코드는 검색 결과를 CSV로 관리하는 Repository입니다.

code_explanation = """
import os
import pandas as pd
from typing import List, Optional
from domain.search_result import SearchResult
from domain.news_article import NewsArticle

class SearchRepository:
    \"\"\"CSV 파일을 사용하여 검색 기록을 관리하는 리포지토리 클래스\"\"\"
    
    # CSV 컬럼 정의 (9개)
    CSV_COLUMNS = [
        "search_key", "search_time", "keyword", "article_index", 
        "title", "url", "snippet", "pub_date", "ai_summary"
    ]

    def __init__(self, csv_path: str):
        self.csv_path = csv_path
        # 데이터 폴더가 없으면 자동 생성!
        os.makedirs(os.path.dirname(self.csv_path), exist_ok=True)

    def load(self) -> pd.DataFrame:
        \"\"\"CSV 파일에서 데이터를 로드\"\"\"
        if not os.path.exists(self.csv_path):
            # 파일 없으면 빈 DataFrame 반환
            return pd.DataFrame(columns=self.CSV_COLUMNS)
        
        try:
            df = pd.read_csv(self.csv_path)
            # 날짜 컬럼을 datetime으로 변환
            if not df.empty:
                df['search_time'] = pd.to_datetime(df['search_time'])
            return df
        except Exception as e:
            print(f"파일 로드 중 오류 발생: {e}")
            return pd.DataFrame(columns=self.CSV_COLUMNS)

    def save(self, search_result: SearchResult) -> bool:
        \"\"\"검색 결과를 CSV 파일에 추가 저장\"\"\"
        try:
            # SearchResult → DataFrame 변환 (to_dataframe 메서드 사용!)
            new_df = search_result.to_dataframe()
            
            # 기존 데이터와 합치기
            existing_df = self.load()
            final_df = pd.concat([existing_df, new_df], ignore_index=True)
            
            # CSV로 저장
            final_df.to_csv(self.csv_path, index=False, encoding='utf-8-sig')
            return True
        except Exception as e:
            print(f"파일 저장 중 오류 발생: {e}")
            return False

    def get_all_keys(self) -> List[str]:
        \"\"\"모든 고유한 search_key를 최신순으로 반환\"\"\"
        df = self.load()
        if df.empty:
            return []
        
        # 최신순 정렬 → 중복 제거 → 리스트
        df_sorted = df.sort_values(by="search_time", ascending=False)
        keys = df_sorted["search_key"].drop_duplicates().tolist()
        return keys

    def find_by_key(self, search_key: str) -> Optional[SearchResult]:
        \"\"\"search_key로 검색 결과를 찾아 SearchResult로 복원\"\"\"
        df = self.load()
        result_df = df[df["search_key"] == search_key]
        if result_df.empty:
            return None
        
        # DataFrame → SearchResult 복원
        first_row = result_df.iloc[0]
        articles = []
        for _, row in result_df.iterrows():
            articles.append(NewsArticle(
                title=row["title"],
                url=row["url"],
                snippet=row["snippet"],
                pub_date=row.get("pub_date")
            ))
        
        return SearchResult(
            search_key=search_key,
            search_time=first_row["search_time"],
            keyword=first_row["keyword"],
            articles=articles,
            ai_summary=first_row["ai_summary"]
        )

    def get_all_as_csv(self) -> str:
        \"\"\"전체 데이터를 CSV 문자열로 반환 (다운로드용)\"\"\"
        df = self.load()
        return df.to_csv(index=False, encoding='utf-8-sig')
"""

print("=== SearchRepository 코드 구조 ===")
print(code_explanation)

print("\n💡 핵심 포인트:")
print("1. CSV_COLUMNS: 저장할 컬럼 미리 정의")
print("2. load(): 파일 없으면 빈 DataFrame 반환 (에러 방지)")
print("3. save(): SearchResult.to_dataframe()으로 변환 후 저장")
print("4. find_by_key(): DataFrame → SearchResult 복원")
print("5. get_all_keys(): 중복 제거된 키 목록 (최신순)")
````

class DataRepository:
    def __init__(self, csv_path: str):
        self.csv_path = csv_path
        os.makedirs(os.path.dirname(csv_path), exist_ok=True)
    
    def load(self) -> pd.DataFrame:
        if not os.path.exists(self.csv_path):
            return pd.DataFrame(columns=[...])
        return pd.read_csv(self.csv_path)
    
    def save(self, data) -> bool:
        new_df = data.to_dataframe()
        df = pd.concat([self.load(), new_df], ignore_index=True)
        df.to_csv(self.csv_path, index=False)
        return True
