## 📚 패키지 관리자(Package Manager)란?

### 이게 뭔가요?

패키지 관리자는 **앱스토어**와 비슷합니다!

- 스마트폰에서 앱을 설치할 때 앱스토어를 사용하죠?
- 파이썬에서 필요한 라이브러리(다른 사람이 만든 코드 모음)를 설치할 때 **패키지 관리자**를 사용합니다
- 예: `pip`, `conda`, 그리고 우리가 사용할 `uv`

### 왜 필요한가요?

만약 패키지 관리자가 없다면...
- 필요한 코드를 일일이 인터넷에서 찾아 다운로드해야 합니다
- 어떤 버전을 받아야 하는지 모릅니다
- 의존성(A를 쓰려면 B가 필요함) 문제를 직접 해결해야 합니다

패키지 관리자가 이 모든 것을 자동으로 해결해줍니다!

### 실행해서 확인해봅시다!
아래 코드를 실행하면 현재 설치된 패키지 목록을 볼 수 있습니다.

````python
# ============================================
# 🔍 현재 설치된 패키지 확인하기
# ============================================
# 이 코드를 실행하면 현재 환경에 설치된 패키지들을 볼 수 있습니다.

# pip list 명령어와 같은 역할 (설치된 패키지 목록)
import pkg_resources  # <- 설치된 패키지 정보를 가져오는 모듈

# 설치된 패키지 개수 확인
installed_packages = list(pkg_resources.working_set)  # 모든 패키지를 리스트로
print(f"현재 설치된 패키지 개수: {len(installed_packages)}개")  # 출력: 숫자

# 처음 10개만 출력해봅시다
print("\n처음 10개 패키지:")
for pkg in installed_packages[:10]:  # 처음 10개만 반복
    print(f"  - {pkg.project_name}: {pkg.version}")  # 이름: 버전 형태로 출력

# 💡 실행 결과를 보면서 이해해보세요!
# - 여러분의 환경에 이미 많은 패키지가 설치되어 있습니다
# - 각 패키지에는 버전 번호가 있습니다 (예: 1.24.0)
````

### 📝 핵심 정리

| 용어 | 설명 |
|------|------|
| **패키지** | 다른 사람이 만든 코드 모음 (라이브러리) |
| **패키지 관리자** | 패키지를 설치/삭제/관리하는 도구 |
| **의존성** | A 패키지가 동작하려면 B 패키지가 필요한 관계 |

💡 **기억하세요**: 패키지 관리자는 프로그래밍의 "앱스토어"입니다!

## 📚 uv란 무엇인가요?

### 이게 뭔가요?

`uv`는 **Rust로 만든 매우 빠른 파이썬 패키지 관리자**입니다.

비유하자면:
- `pip`이 **일반 택배**라면
- `uv`는 **로켓 배송**입니다! 🚀

같은 일을 하지만 uv가 10~100배 더 빠릅니다.

### 왜 필요한가요?

1. **속도**: 패키지 설치가 매우 빠릅니다
2. **통합 기능**: 가상환경 생성 + 패키지 설치 + 버전 잠금을 한 도구에서
3. **안정성**: 의존성 잠금 파일(`uv.lock`)로 재현 가능한 환경 제공

### 실행해서 확인해봅시다!
uv의 주요 명령어들을 살펴봅시다.

````python
# ============================================
# 🔍 uv 주요 명령어 살펴보기
# ============================================
# 이 셀은 실행하지 않고 설명만 읽어주세요!
# 실제 명령어는 터미널에서 실행해야 합니다.

# uv 주요 명령어 설명
uv_commands = {
    "uv init 프로젝트명": "새 프로젝트 시작 (폴더 생성 + pyproject.toml 생성)",
    "uv venv": "가상환경 생성 (.venv 폴더 만들기)",
    "uv venv --python 3.11": "특정 파이썬 버전으로 가상환경 생성",
    "uv add 패키지명": "패키지 설치 (pip install과 같은 역할)",
    "uv add streamlit pandas": "여러 패키지 한번에 설치",
    "uv pip list": "설치된 패키지 목록 확인",
    "uv sync": "uv.lock 파일 기준으로 패키지 동기화",
}

print("=" * 50)
print("📋 uv 주요 명령어 정리")
print("=" * 50)

for command, description in uv_commands.items():
    print(f"\n🔹 {command}")
    print(f"   └─ {description}")

print("\n" + "=" * 50)
print("💡 이 명령어들은 터미널에서 실행합니다!")
print("=" * 50)
````

### 📝 핵심 정리

| 명령어 | 하는 일 |
|--------|--------|
| `uv init` | 프로젝트 초기화 |
| `uv venv` | 가상환경 생성 |
| `uv add` | 패키지 설치 |
| `uv sync` | 의존성 동기화 |

💡 **기억하세요**: uv는 pip보다 훨씬 빠른 현대적인 패키지 관리자입니다!

## 📚 가상환경(Virtual Environment)이란?

### 이게 뭔가요?

가상환경은 **프로젝트별 독립된 방**과 같습니다!

비유:
- 집 전체(컴퓨터)에 물건(패키지)을 두면 어질러지겠죠?
- 각 프로젝트마다 **개인 방(가상환경)**을 만들어 필요한 물건만 둡니다
- A 프로젝트 방에는 pandas 1.0, B 프로젝트 방에는 pandas 2.0 가능!

### 왜 필요한가요?

| 상황 | 가상환경 없이 | 가상환경 있으면 |
|------|--------------|----------------|
| A 프로젝트: pandas 1.0 필요 | ❌ 충돌! | ✅ A방에 1.0 설치 |
| B 프로젝트: pandas 2.0 필요 | ❌ 하나만 가능 | ✅ B방에 2.0 설치 |

### 실행해서 확인해봅시다!
현재 가상환경 정보를 확인해봅시다.

````python
# ============================================
# 🔍 가상환경 정보 확인하기
# ============================================
# 이 코드를 실행하면 현재 파이썬 실행 환경을 확인할 수 있습니다.

import sys  # <- 파이썬 시스템 정보를 가져오는 모듈
import os   # <- 운영체제 관련 기능을 제공하는 모듈

# 현재 파이썬 실행 파일 경로
print("🐍 파이썬 실행 파일 위치:")
print(f"   {sys.executable}")  # 출력 예: /path/to/.venv/bin/python

# 가상환경 확인
print("\n📂 가상환경 확인:")
if '.venv' in sys.executable or 'venv' in sys.executable:
    print("   ✅ 가상환경 안에서 실행 중입니다!")
else:
    print("   ⚠️ 가상환경이 아닌 시스템 파이썬입니다.")

# 파이썬 버전 확인
print(f"\n🔢 파이썬 버전: {sys.version}")

# 💡 실행 결과를 보면서 이해해보세요!
# - 경로에 .venv가 있으면 가상환경 안에서 실행 중입니다
# - 프로젝트마다 다른 가상환경을 사용할 수 있습니다
````

### 📝 핵심 정리

| 용어 | 설명 |
|------|------|
| **가상환경** | 프로젝트별로 분리된 파이썬 환경 |
| `.venv` 폴더 | 가상환경이 저장되는 폴더 |
| `activate` | 가상환경을 활성화하는 명령어 |

💡 **기억하세요**: 가상환경은 프로젝트마다 깨끗한 독립 공간을 만들어줍니다!

## 📚 환경변수(Environment Variable)와 .env 파일

### 이게 뭔가요?

환경변수는 **비밀 금고**와 같습니다!

비유:
- 집 열쇠를 현관문 아래 숨기면? 🚫 위험!
- **금고(환경변수)**에 보관하고 필요할 때 꺼내쓰면? ✅ 안전!

**API 키**(외부 서비스 이용 비밀번호)를 코드에 직접 쓰면 위험합니다:
```python
# ❌ 나쁜 예시 - 절대 이렇게 하지 마세요!
api_key = "sk-abc123실제키"  # 코드를 공유하면 키도 노출!
```

### 왜 필요한가요?

1. **보안**: API 키가 코드에 노출되지 않음
2. **유연성**: 개발/운영 환경마다 다른 설정 가능
3. **편의성**: 한 곳에서 모든 설정 관리

### 실행해서 확인해봅시다!
환경변수를 읽고 쓰는 방법을 알아봅시다.

````python
# ============================================
# 🔍 환경변수 기본 사용법 확인하기
# ============================================
# 이 코드를 실행하면 환경변수를 읽고 쓰는 방법을 볼 수 있습니다.

import os  # <- 운영체제와 상호작용하는 모듈

# 방법 1: 환경변수 읽기 (없으면 None)
user = os.getenv("USER")  # 현재 사용자 이름 (Linux/Mac)
print(f"현재 사용자: {user}")  # 출력: 사용자 이름 또는 None

# 방법 2: 환경변수 읽기 (없으면 기본값 사용)
my_setting = os.getenv("MY_SETTING", "기본값입니다")
print(f"MY_SETTING 값: {my_setting}")  # 출력: 기본값입니다

# 방법 3: 환경변수 설정하기 (이 프로그램 내에서만 유효)
os.environ["MY_TEST_VAR"] = "안녕하세요!"  # 환경변수 설정
print(f"MY_TEST_VAR 값: {os.getenv('MY_TEST_VAR')}")  # 출력: 안녕하세요!

# 💡 실행 결과를 보면서 이해해보세요!
# - os.getenv()로 환경변수를 안전하게 읽을 수 있습니다
# - 없는 환경변수는 None 또는 기본값을 반환합니다
````

````python
# ============================================
# 🔍 .env 파일 사용법 확인하기
# ============================================
# python-dotenv 라이브러리를 사용하면 .env 파일에서 환경변수를 로드합니다.

# .env 파일 예시 (실제 파일 내용)
env_file_example = """
# ========================================
# .env 파일 예시
# ========================================

# Tavily API (검색 서비스)
TAVILY_API_KEY=tvly-xxxxxxxxxxxx

# Google Gemini API (AI 서비스)
GEMINI_API_KEY=your_gemini_api_key

# 데이터 저장 경로
CSV_PATH=data/search_history.csv
"""

print(env_file_example)

print("\n💡 사용 방법:")
print("1. .env.example 파일을 .env로 복사")
print("2. 실제 API 키를 입력")
print("3. .env 파일은 절대 Git에 올리지 않음! (.gitignore에 추가)")
````

````python
# ============================================
# 🔍 dotenv로 .env 파일 로드하기
# ============================================
# load_dotenv() 함수가 .env 파일의 내용을 환경변수로 로드합니다.

from dotenv import load_dotenv  # <- .env 파일을 읽는 라이브러리
import os

# .env 파일 로드 시도
result = load_dotenv()  # 현재 폴더에서 .env 파일을 찾아 로드

if result:
    print("✅ .env 파일을 성공적으로 로드했습니다!")
else:
    print("⚠️ .env 파일을 찾지 못했습니다. (정상적인 상황일 수 있음)")

# 로드된 환경변수 확인 (실제 키는 출력하지 않음!)
gemini_key = os.getenv("GEMINI_API_KEY")
if gemini_key:
    # 보안을 위해 처음 4글자만 표시
    print(f"\nGEMINI_API_KEY: {gemini_key[:4]}... (설정됨)")
else:
    print("\nGEMINI_API_KEY: 설정되지 않음")

# 💡 실행 결과를 보면서 이해해보세요!
# - load_dotenv()가 .env 파일을 읽어 환경변수로 만들어줍니다
# - API 키는 절대 전체를 출력하지 않습니다 (보안!)
````

### 📝 핵심 정리

| 항목 | 설명 |
|------|------|
| `.env` 파일 | API 키 등 비밀 설정을 저장하는 파일 |
| `.env.example` | .env 파일의 템플릿 (공유 가능) |
| `load_dotenv()` | .env 파일을 읽어 환경변수로 로드 |
| `os.getenv()` | 환경변수 값을 안전하게 읽기 |

💡 **기억하세요**: API 키는 코드에 직접 쓰지 말고, .env 파일에 저장하세요!

## 📚 프로젝트 폴더 구조 (레이어드 아키텍처)

### 이게 뭔가요?

폴더 구조는 **회사 조직도**와 같습니다!

비유:
- 모든 직원이 한 방에 있으면? 🚫 혼란!
- **부서(폴더)**별로 나누면? ✅ 체계적!

```
회사(프로젝트)
├── 영업부(services)     → 외부와 소통 (API 호출)
├── 창고(repositories)   → 물건 보관 (데이터 저장)
├── 기획부(domain)       → 핵심 업무 정의 (데이터 모델)
├── 고객센터(components) → 고객 응대 (UI)
└── 총무부(config)       → 회사 설정 관리 (환경설정)
```

### 왜 필요한가요?

1. **찾기 쉬움**: 어디에 뭐가 있는지 바로 알 수 있음
2. **수정 쉬움**: 한 부분을 바꿔도 다른 부분에 영향 없음
3. **협업 쉬움**: 여러 명이 다른 폴더 담당 가능

## 🔎 config/settings.py 분석

이제 배운 개념이 initial_version 프로젝트에서 어떻게 사용되는지 살펴봅시다.

**파일**: `config/settings.py`
**역할**: 환경변수를 로드하고, 필수값이 없으면 친절한 에러 메시지를 보여줌

````python
# ============================================
# 🔎 initial_version의 config/settings.py 코드
# ============================================
# 이 코드는 환경변수를 로드하고 검증하는 Settings 클래스입니다.

# 아래는 실제 프로젝트 코드를 학습용으로 정리한 버전입니다.
# 각 줄의 역할을 주석으로 설명합니다.

import os                     # <- 환경변수를 다루는 모듈
from dotenv import load_dotenv  # <- .env 파일을 읽는 함수

class Settings:  # <- Settings라는 클래스(설계도) 정의
    """애플리케이션의 환경 변수 및 설정을 관리하는 클래스"""
    
    def __init__(self):  # <- 객체가 생성될 때 자동 실행되는 함수
        # 1단계: .env 파일 로드
        load_dotenv()  # <- .env 파일의 내용을 환경변수로 등록
        
        # 2단계: 필수 환경변수 목록 정의
        required_vars = {  # <- 딕셔너리로 변수명: 설명 형태
            "TAVILY_API_KEY": "Tavily API 키가 필요합니다.",
            "GEMINI_API_KEY": "Google Gemini API 키가 필요합니다.",
            "CSV_PATH": "CSV 파일 경로가 필요합니다."
        }
        
        # 3단계: 누락된 변수 검사
        missing_vars = []  # <- 빈 리스트
        for var, description in required_vars.items():  # <- 키, 값 쌍으로 반복
            value = os.getenv(var)  # <- 환경변수 읽기
            if not value:           # <- 값이 없으면
                missing_vars.append(f"- {var}: {description}")  # <- 목록에 추가
        
        # 4단계: 누락된 변수가 있으면 에러 발생
        if missing_vars:
            error_msg = f"\n❌ 환경변수가 설정되지 않았습니다.\n누락된 변수:\n"
            error_msg += "\n".join(missing_vars)  # <- 리스트를 줄바꿈으로 연결
            raise ValueError(error_msg)  # <- 에러 발생시키기
        
        # 5단계: 설정값 저장 (self.변수명으로 어디서든 접근 가능)
        self.TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")
        self.GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
        self.CSV_PATH = os.getenv("CSV_PATH")
        
        # 6단계: 선택적 설정 (기본값 제공)
        self.GEMINI_MODEL = os.getenv("GEMINI_MODEL", "gemini-2.5-flash")

# 💡 프로젝트에서 이 코드가 하는 역할:
# - 앱 시작 시 Settings() 객체를 만들면 자동으로 .env를 읽습니다
# - 필수 설정이 없으면 친절한 에러 메시지와 함께 앱이 종료됩니다
# - settings.TAVILY_API_KEY 처럼 어디서든 설정값에 접근할 수 있습니다

print("✅ Settings 클래스 코드를 이해했습니다!")
print("\n이 클래스는 다음과 같이 사용됩니다:")
print(">>> settings = Settings()")
print(">>> print(settings.GEMINI_MODEL)")
print(">>> # 출력: gemini-2.5-flash")
````
