````python
# ============================================
# ğŸ” í•¨ìˆ˜ì—ì„œ íƒ€ì… íŒíŠ¸ ì‚¬ìš©
# ============================================
# í•¨ìˆ˜ì˜ ë§¤ê°œë³€ìˆ˜ì™€ ë°˜í™˜ê°’ì— íƒ€ì… íŒíŠ¸ë¥¼ ë¶™ì…ë‹ˆë‹¤.

from typing import List, Optional

def calculate_average(scores: List[int]) -> float:
    """
    ì ìˆ˜ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°›ì•„ í‰ê· ì„ ê³„ì‚°í•©ë‹ˆë‹¤.
    
    Args:
        scores: ì •ìˆ˜ë¡œ ëœ ì ìˆ˜ ë¦¬ìŠ¤íŠ¸
    Returns:
        í‰ê· ê°’ (ì†Œìˆ˜ì )
    """
    if not scores:  # ë¹ˆ ë¦¬ìŠ¤íŠ¸ë©´
        return 0.0
    return sum(scores) / len(scores)

def find_max_score(scores: List[int]) -> Optional[int]:
    """
    ìµœê³  ì ìˆ˜ë¥¼ ì°¾ìŠµë‹ˆë‹¤. ë¹ˆ ë¦¬ìŠ¤íŠ¸ë©´ None ë°˜í™˜.
    """
    if not scores:
        return None  # Optionalì´ë¯€ë¡œ None ë°˜í™˜ ê°€ëŠ¥
    return max(scores)

# í•¨ìˆ˜ ì‚¬ìš©
my_scores = [85, 90, 78, 92, 88]

avg = calculate_average(my_scores)
print(f"ì ìˆ˜: {my_scores}")
print(f"í‰ê· : {avg}")

max_score = find_max_score(my_scores)
print(f"ìµœê³ ì : {max_score}")

# ë¹ˆ ë¦¬ìŠ¤íŠ¸ í…ŒìŠ¤íŠ¸
empty_scores: List[int] = []
print(f"\në¹ˆ ë¦¬ìŠ¤íŠ¸ í‰ê· : {calculate_average(empty_scores)}")
print(f"ë¹ˆ ë¦¬ìŠ¤íŠ¸ ìµœê³ ì : {find_max_score(empty_scores)}")

# ğŸ’¡ -> float, -> Optional[int] ëŠ” ë°˜í™˜ íƒ€ì…ì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.
````

## ğŸ” domain/search_result.py ë¶„ì„

**íŒŒì¼**: `domain/search_result.py`

````python
# ============================================
# ğŸ” initial_versionì˜ domain/search_result.py ì½”ë“œ
# ============================================
# ì´ ì½”ë“œëŠ” ê²€ìƒ‰ ê²°ê³¼ ì „ì²´ë¥¼ ë‹´ëŠ” ë°ì´í„° í´ë˜ìŠ¤ì…ë‹ˆë‹¤.

from dataclasses import dataclass
from datetime import datetime
from typing import List
import pandas as pd

# NewsArticle í´ë˜ìŠ¤ëŠ” ìœ„ì—ì„œ ì´ë¯¸ ì •ì˜ë¨

@dataclass
class SearchResult:
    """ê²€ìƒ‰ ê²°ê³¼ ë° AI ìš”ì•½ ì •ë³´ë¥¼ ë‹´ëŠ” ë°ì´í„° í´ë˜ìŠ¤"""
    search_key: str          # í˜•ì‹: "í‚¤ì›Œë“œ-yyyymmddhhmm" (ê³ ìœ  ì‹ë³„ì)
    search_time: datetime    # ê²€ìƒ‰ ì‹¤í–‰ ì‹œê°„
    keyword: str             # ê²€ìƒ‰ í‚¤ì›Œë“œ
    articles: List[NewsArticle]  # ë‰´ìŠ¤ ê¸°ì‚¬ ë¦¬ìŠ¤íŠ¸ (ì—¬ëŸ¬ ê°œ!)
    ai_summary: str          # AI ìš”ì•½ ê²°ê³¼

    def to_dataframe(self) -> pd.DataFrame:
        """
        ê²€ìƒ‰ ê²°ê³¼ë¥¼ Pandas DataFrameìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
        ê¸°ì‚¬ 1ê±´ì´ 1í–‰ì´ ë˜ëŠ” Long formatìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
        """
        data = []  # ë³€í™˜ëœ ë°ì´í„°ë¥¼ ë‹´ì„ ë¦¬ìŠ¤íŠ¸
        
        for i, article in enumerate(self.articles, 1):  # 1ë¶€í„° ë²ˆí˜¸ ë§¤ê¸°ê¸°
            data.append({  # ê° ê¸°ì‚¬ë¥¼ ë”•ì…”ë„ˆë¦¬ë¡œ
                "search_key": self.search_key,
                "search_time": self.search_time,
                "keyword": self.keyword,
                "article_index": i,
                "title": article.title,
                "url": article.url,
                "snippet": article.snippet,
                "pub_date": article.pub_date,
                "ai_summary": self.ai_summary
            })
        
        return pd.DataFrame(data)  # DataFrameìœ¼ë¡œ ë³€í™˜

# ğŸ’¡ í”„ë¡œì íŠ¸ì—ì„œ ì´ ì½”ë“œê°€ í•˜ëŠ” ì—­í• :
# - ê²€ìƒ‰ ê²°ê³¼ë¥¼ í•˜ë‚˜ì˜ ê°ì²´ë¡œ ë¬¶ì–´ ê´€ë¦¬
# - to_dataframe()ìœ¼ë¡œ CSV ì €ì¥ì— ì í•©í•œ í˜•íƒœë¡œ ë³€í™˜
# - search_keyë¡œ ê³¼ê±° ê²€ìƒ‰ ê¸°ë¡ ì¡°íšŒ ê°€ëŠ¥

print("âœ… SearchResult í´ë˜ìŠ¤ ì½”ë“œë¥¼ ì´í•´í–ˆìŠµë‹ˆë‹¤!")
````
