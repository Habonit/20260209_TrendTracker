## 📚 클래스(class)란?

### 이게 뭔가요?

클래스는 **붕어빵 틀**과 같습니다! 🐟

| 비유 | 프로그래밍 |
|------|------------|
| 붕어빵 틀 | 클래스(class) - 모양을 정의 |
| 붕어빵 | 인스턴스(instance) - 틀로 만든 실제 결과물 |
| 팥/슈크림 | 속성(attribute) - 각 붕어빵의 특징 |

하나의 틀(클래스)로 여러 개의 붕어빵(인스턴스)을 찍어낼 수 있습니다!

### 왜 필요한가요?

뉴스 기사 100개를 다룬다고 상상해보세요:

**클래스 없이:**
```python
title1 = "첫 번째 기사"
url1 = "https://..."
snippet1 = "내용..."

title2 = "두 번째 기사"
url2 = "https://..."
snippet2 = "내용..."
# 변수가 300개!!! 😱
```

**클래스 사용:**
```python
article1 = NewsArticle(title="첫 번째", url="...", snippet="...")
article2 = NewsArticle(title="두 번째", url="...", snippet="...")
# 깔끔하게 관리! 😊
```

### 실행해서 확인해봅시다!
클래스를 만들고 사용하는 방법을 직접 실행해봅시다.

````python
# ============================================
# 🔍 클래스 기본 사용법 확인하기
# ============================================
# 붕어빵 틀(클래스)을 만들고, 실제 붕어빵(인스턴스)을 찍어봅시다!

# 붕어빵 틀(클래스) 정의
class 붕어빵:  # <- class 키워드로 클래스 정의
    def __init__(self, 속재료):  # <- 붕어빵을 만들 때 호출되는 특별한 함수
        self.속재료 = 속재료     # <- self.변수 = 값으로 속성 저장

# 틀로 붕어빵 2개 만들기 (인스턴스 생성)
팥붕 = 붕어빵("팥")           # <- 팥이 든 붕어빵
슈크림붕 = 붕어빵("슈크림")    # <- 슈크림이 든 붕어빵

# 각 붕어빵의 속재료 확인
print(f"첫 번째 붕어빵: {팥붕.속재료}")      # 출력: 팥
print(f"두 번째 붕어빵: {슈크림붕.속재료}")  # 출력: 슈크림

# 💡 실행 결과를 보면서 이해해보세요!
# - 같은 틀(class)로 다른 붕어빵(인스턴스)을 만들 수 있습니다
# - 각 붕어빵은 자신만의 속재료(속성)를 가집니다
# - .속재료 처럼 점(.)으로 속성에 접근합니다
````

````python
# ============================================
# 🔍 실제 프로그래밍에서의 클래스 예시
# ============================================
# 학생 정보를 관리하는 클래스를 만들어봅시다.

class Student:  # <- 학생 클래스 정의
    def __init__(self, name, age, grade):  # <- 생성자: 학생 만들 때 필요한 정보
        self.name = name    # <- 이름 저장
        self.age = age      # <- 나이 저장
        self.grade = grade  # <- 학년 저장
    
    def introduce(self):  # <- 메서드: 클래스 안의 함수
        return f"안녕하세요! {self.grade}학년 {self.name}입니다."

# 학생 인스턴스 생성
student1 = Student("김철수", 20, 2)
student2 = Student("이영희", 21, 3)

# 정보 확인
print(f"학생1 이름: {student1.name}, 나이: {student1.age}")
print(f"학생2 이름: {student2.name}, 나이: {student2.age}")

# 메서드 호출
print(student1.introduce())  # 출력: 안녕하세요! 2학년 김철수입니다.
print(student2.introduce())  # 출력: 안녕하세요! 3학년 이영희입니다.

# 💡 메서드(introduce)는 클래스 안에 정의된 함수입니다.
# - self를 통해 인스턴스의 속성(name, grade)에 접근할 수 있습니다.
````

### 📝 핵심 정리

| 용어 | 설명 | 비유 |
|------|------|------|
| `class` | 객체의 설계도 | 붕어빵 틀 |
| 인스턴스 | 클래스로 만든 실제 객체 | 붕어빵 |
| `__init__` | 인스턴스 생성 시 호출되는 함수 | 틀에 반죽 넣기 |
| `self` | 자기 자신(인스턴스)을 가리킴 | "나" |
| 속성 | 인스턴스가 가진 데이터 | 속재료 |
| 메서드 | 클래스 안에 정의된 함수 | 행동 |

💡 **기억하세요**: 클래스는 "같은 구조의 데이터를 여러 개 만들 때" 사용합니다!

## 📚 @dataclass란?

### 이게 뭔가요?

`@dataclass`는 **자동 붕어빵 기계**와 같습니다! 🤖

일반 클래스는 `__init__`을 직접 작성해야 하지만,
`@dataclass`를 붙이면 **자동으로 만들어줍니다!**

```python
# 😰 일반 클래스 - 직접 다 작성
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

# 😊 dataclass - 자동 생성!
@dataclass
class Person:
    name: str
    age: int
```

### 왜 필요한가요?

1. **코드가 짧아짐**: `__init__` 안 써도 됨
2. **자동 기능**: 비교, 출력 등이 자동으로 됨
3. **실수 방지**: 타입을 명시해서 실수 줄임

### 실행해서 확인해봅시다!
일반 클래스와 dataclass를 비교해봅시다.

````python
# ============================================
# 🔍 일반 클래스 vs @dataclass 비교
# ============================================
# 같은 기능을 하는 두 클래스를 비교해봅시다.

from dataclasses import dataclass  # <- dataclass를 사용하기 위해 import

# 방법 1: 일반 클래스 (코드가 길다)
class PersonNormal:
    def __init__(self, name, age, email):
        self.name = name
        self.age = age
        self.email = email
    
    def __repr__(self):  # 출력 형식 정의 (없으면 못생긴 출력)
        return f"PersonNormal(name='{self.name}', age={self.age}, email='{self.email}')"

# 방법 2: @dataclass (코드가 짧다!)
@dataclass
class PersonDataclass:
    name: str   # 변수명: 타입 형태로 작성
    age: int
    email: str

# 두 방식으로 객체 생성
person1 = PersonNormal("김철수", 25, "kim@email.com")
person2 = PersonDataclass("이영희", 23, "lee@email.com")

# 출력 비교
print("일반 클래스:")
print(f"  {person1}")

print("\n@dataclass:")
print(f"  {person2}")

# 💡 @dataclass는 __init__과 __repr__을 자동으로 만들어줍니다!
````

````python
# ============================================
# 🔍 @dataclass의 편리한 기능들
# ============================================
# dataclass가 자동으로 제공하는 기능들을 살펴봅시다.

from dataclasses import dataclass

@dataclass
class Book:
    title: str
    author: str
    price: int

# 인스턴스 생성
book1 = Book("파이썬 입문", "홍길동", 25000)
book2 = Book("파이썬 입문", "홍길동", 25000)  # book1과 같은 내용
book3 = Book("자바 입문", "김철수", 30000)

# 기능 1: 자동 출력 (__repr__)
print("1. 자동 출력:")
print(f"   {book1}")

# 기능 2: 자동 비교 (__eq__)
print("\n2. 자동 비교:")
print(f"   book1 == book2: {book1 == book2}")  # True (내용이 같음)
print(f"   book1 == book3: {book1 == book3}")  # False (내용이 다름)

# 기능 3: 속성 접근
print("\n3. 속성 접근:")
print(f"   제목: {book1.title}")
print(f"   저자: {book1.author}")
print(f"   가격: {book1.price}원")

# 💡 dataclass는 __init__, __repr__, __eq__ 등을 자동 생성합니다!
````

````python
# ============================================
# 🔍 @dataclass에서 기본값 사용하기
# ============================================
# 일부 속성에 기본값을 지정할 수 있습니다.

from dataclasses import dataclass
from typing import Optional  # <- 값이 있거나 None일 수 있음을 표시

@dataclass
class Product:
    name: str                      # 필수값 (기본값 없음)
    price: int                     # 필수값
    category: str = "기타"         # 선택값 (기본값: "기타")
    discount: int = 0              # 선택값 (기본값: 0)
    description: Optional[str] = None  # 선택값 (None 허용)

# 필수값만 전달
product1 = Product("노트북", 1500000)
print(f"상품1: {product1}")
# 출력: Product(name='노트북', price=1500000, category='기타', discount=0, description=None)

# 일부 선택값 전달
product2 = Product("마우스", 50000, category="전자기기", discount=10)
print(f"상품2: {product2}")
# 출력: Product(name='마우스', price=50000, category='전자기기', discount=10, description=None)

# 💡 주의: 기본값이 있는 속성은 반드시 기본값 없는 속성 뒤에 와야 합니다!
````

### 📝 핵심 정리

| 항목 | 일반 클래스 | @dataclass |
|------|------------|------------|
| `__init__` | 직접 작성 | 자동 생성 |
| `__repr__` | 직접 작성 | 자동 생성 |
| `__eq__` | 직접 작성 | 자동 생성 |
| 코드 길이 | 길다 | 짧다 |
| 용도 | 복잡한 로직 | 데이터 저장 |

💡 **기억하세요**: 데이터만 저장하는 클래스는 `@dataclass`를 사용하세요!

## 📚 타입 힌트(Type Hint)란?

### 이게 뭔가요?

타입 힌트는 **택배 상자의 라벨**과 같습니다! 📦

- 상자에 "깨지는 물건"이라고 써있으면 조심히 다루죠?
- 변수에 `str`, `int`라고 표시하면 어떤 값을 넣어야 하는지 알 수 있습니다.

```python
# 타입 힌트 없이
def greet(name):
    return f"안녕, {name}!"

# 타입 힌트 있으면
def greet(name: str) -> str:
    return f"안녕, {name}!"
# name은 문자열(str)이고, 반환값도 문자열(str)!
```

### 왜 필요한가요?

1. **실수 방지**: 잘못된 타입을 넣으면 IDE가 경고
2. **자동 완성**: IDE가 어떤 속성/메서드가 있는지 알려줌
3. **문서 역할**: 코드만 봐도 어떤 값을 기대하는지 알 수 있음

### 실행해서 확인해봅시다!
타입 힌트의 여러 사용법을 살펴봅시다.

````python
# ============================================
# 🔍 기본 타입 힌트 사용법
# ============================================
# 파이썬의 기본 타입들을 알아봅시다.

# 기본 타입
name: str = "홍길동"      # 문자열
age: int = 25             # 정수
height: float = 175.5     # 실수(소수점)
is_student: bool = True   # 참/거짓

print("=== 기본 타입 ===")
print(f"name: {name} (타입: {type(name).__name__})")
print(f"age: {age} (타입: {type(age).__name__})")
print(f"height: {height} (타입: {type(height).__name__})")
print(f"is_student: {is_student} (타입: {type(is_student).__name__})")
````

````python
# ============================================
# 🔍 컬렉션 타입 힌트
# ============================================
# 리스트, 딕셔너리 등 여러 값을 담는 타입

from typing import List, Dict, Tuple, Optional

# List: 같은 타입의 여러 값
scores: List[int] = [90, 85, 88, 92]
names: List[str] = ["철수", "영희", "민수"]

print("=== 리스트 타입 ===")
print(f"scores: {scores}")
print(f"names: {names}")

# Dict: 키-값 쌍
grades: Dict[str, int] = {"철수": 90, "영희": 85}
print(f"\ngrades: {grades}")

# Tuple: 고정된 개수의 값들
point: Tuple[int, int] = (10, 20)  # x, y 좌표
print(f"point: {point}")

# Optional: 값이 있거나 None
nickname: Optional[str] = None  # 별명은 없을 수도 있음
nickname = "파이썬마스터"        # 나중에 값 설정 가능
print(f"\nnickname: {nickname}")
````

````python
# ============================================
# 🔍 함수에서 타입 힌트 사용
# ============================================
# 함수의 매개변수와 반환값에 타입 힌트를 붙입니다.

from typing import List, Optional

def calculate_average(scores: List[int]) -> float:
    """
    점수 리스트를 받아 평균을 계산합니다.
    
    Args:
        scores: 정수로 된 점수 리스트
    Returns:
        평균값 (소수점)
    """
    if not scores:  # 빈 리스트면
        return 0.0
    return sum(scores) / len(scores)

def find_max_score(scores: List[int]) -> Optional[int]:
    """
    최고 점수를 찾습니다. 빈 리스트면 None 반환.
    """
    if not scores:
        return None  # Optional이므로 None 반환 가능
    return max(scores)

# 함수 사용
my_scores = [85, 90, 78, 92, 88]

avg = calculate_average(my_scores)
print(f"점수: {my_scores}")
print(f"평균: {avg}")

max_score = find_max_score(my_scores)
print(f"최고점: {max_score}")

# 빈 리스트 테스트
empty_scores: List[int] = []
print(f"\n빈 리스트 평균: {calculate_average(empty_scores)}")
print(f"빈 리스트 최고점: {find_max_score(empty_scores)}")

# 💡 -> float, -> Optional[int] 는 반환 타입을 나타냅니다.
````

### 📝 핵심 정리

| 타입 | 설명 | 예시 |
|------|------|------|
| `str` | 문자열 | `"안녕"` |
| `int` | 정수 | `42` |
| `float` | 실수 | `3.14` |
| `bool` | 참/거짓 | `True` |
| `List[X]` | X 타입의 리스트 | `[1, 2, 3]` |
| `Dict[K, V]` | K-V 쌍의 딕셔너리 | `{"a": 1}` |
| `Optional[X]` | X 또는 None | `None` 또는 값 |

💡 **기억하세요**: 타입 힌트는 "이 변수에는 이런 값이 들어갑니다"라는 라벨입니다!

## 📚 도메인 모델이란?

### 이게 뭔가요?

도메인 모델은 **프로그램이 다루는 핵심 개념을 코드로 표현**한 것입니다.

비유:
- 도서관 프로그램 → `Book`, `Member`, `Loan` 모델
- 쇼핑몰 프로그램 → `Product`, `Order`, `Customer` 모델
- **뉴스 검색 프로그램** → `NewsArticle`, `SearchResult` 모델

### 왜 필요한가요?

1. **명확한 구조**: "뉴스 기사"가 무엇으로 구성되는지 정의
2. **재사용**: 한 번 정의하면 프로그램 전체에서 사용
3. **유지보수**: 변경이 필요하면 한 곳만 수정

### 실행해서 확인해봅시다!
도메인 모델의 예시를 살펴봅시다.

## 🔎 domain/search_result.py 분석

**파일**: `domain/search_result.py`
**역할**: 검색 결과 전체(여러 기사 + AI 요약)를 담는 데이터 클래스

````python
# ============================================
# 🔎 initial_version의 domain/search_result.py 코드
# ============================================
# 이 코드는 검색 결과 전체를 담는 데이터 클래스입니다.

from dataclasses import dataclass
from datetime import datetime
from typing import List
import pandas as pd

# NewsArticle 클래스는 위에서 이미 정의됨

@dataclass
class SearchResult:
    """검색 결과 및 AI 요약 정보를 담는 데이터 클래스"""
    search_key: str          # 형식: "키워드-yyyymmddhhmm" (고유 식별자)
    search_time: datetime    # 검색 실행 시간
    keyword: str             # 검색 키워드
    articles: List[NewsArticle]  # 뉴스 기사 리스트 (여러 개!)
    ai_summary: str          # AI 요약 결과

    def to_dataframe(self) -> pd.DataFrame:
        """
        검색 결과를 Pandas DataFrame으로 변환합니다.
        기사 1건이 1행이 되는 Long format으로 변환합니다.
        """
        data = []  # 변환된 데이터를 담을 리스트
        
        for i, article in enumerate(self.articles, 1):  # 1부터 번호 매기기
            data.append({  # 각 기사를 딕셔너리로
                "search_key": self.search_key,
                "search_time": self.search_time,
                "keyword": self.keyword,
                "article_index": i,
                "title": article.title,
                "url": article.url,
                "snippet": article.snippet,
                "pub_date": article.pub_date,
                "ai_summary": self.ai_summary
            })
        
        return pd.DataFrame(data)  # DataFrame으로 변환

# 💡 프로젝트에서 이 코드가 하는 역할:
# - 검색 결과를 하나의 객체로 묶어 관리
# - to_dataframe()으로 CSV 저장에 적합한 형태로 변환
# - search_key로 과거 검색 기록 조회 가능

print("✅ SearchResult 클래스 코드를 이해했습니다!")
````

````python
# ============================================
# 🔎 SearchResult 실제 사용 예시
# ============================================
# 검색 결과를 만들고 DataFrame으로 변환해봅시다.

# 기사 3개 생성
articles = [
    NewsArticle("AI 뉴스 1", "https://1.com", "첫 번째 기사", "2024-01-20"),
    NewsArticle("AI 뉴스 2", "https://2.com", "두 번째 기사", "2024-01-19"),
    NewsArticle("AI 뉴스 3", "https://3.com", "세 번째 기사")
]

# SearchResult 생성
result = SearchResult(
    search_key="인공지능-202401201430",
    search_time=datetime.now(),
    keyword="인공지능",
    articles=articles,
    ai_summary="AI는 우리 생활을 크게 변화시키고 있습니다."
)

print("=== SearchResult 정보 ===")
print(f"검색키: {result.search_key}")
print(f"키워드: {result.keyword}")
print(f"기사 수: {len(result.articles)}개")

# DataFrame으로 변환
df = result.to_dataframe()
print("\n=== DataFrame 변환 결과 ===")
print(df[["article_index", "title", "pub_date"]])

# 💡 to_dataframe()으로 CSV 저장에 적합한 형태가 됩니다!
````

## 🔎 utils/key_generator.py 분석

**파일**: `utils/key_generator.py`
**역할**: 검색 결과의 고유 식별자(search_key)를 생성

````python
# ============================================
# 🔎 initial_version의 utils/key_generator.py 코드
# ============================================
# 이 코드는 검색 결과의 고유 식별자를 생성합니다.

from datetime import datetime

def generate_search_key(keyword: str) -> str:
    """
    현재 시간을 기준으로 "키워드-yyyymmddhhmm" 형식의 검색 키를 생성합니다.
    
    Args:
        keyword: 검색 키워드
    Returns:
        "키워드-yyyymmddhhmm" 형식의 문자열
    """
    now = datetime.now()  # 현재 시간
    timestamp = now.strftime("%Y%m%d%H%M")  # 202401201430 형식
    return f"{keyword}-{timestamp}"  # 키워드-타임스탬프

# 💡 프로젝트에서 이 코드가 하는 역할:
# - 각 검색마다 고유한 키 생성
# - 이 키로 과거 검색 기록을 찾을 수 있음

# 실행 예시
key1 = generate_search_key("인공지능")
print(f"생성된 키: {key1}")
# 출력 예: 인공지능-202401201430
````

# 📋 정리

## 이번 Phase에서 배운 것

| 개념 | 한 줄 요약 |
|------|-----------|
| **클래스(class)** | 같은 구조의 데이터를 여러 개 만드는 틀 (붕어빵 틀) |
| **인스턴스** | 클래스로 만든 실제 객체 (붕어빵) |
| **@dataclass** | 데이터 저장용 클래스를 간편하게 만드는 도구 |
| **타입 힌트** | 변수/함수에 어떤 타입이 오는지 표시 (라벨) |
| **Optional** | 값이 있거나 None일 수 있음을 표시 |
| **도메인 모델** | 프로그램이 다루는 핵심 개념을 코드로 표현 |

## 핵심 코드 패턴

```python
# dataclass로 도메인 모델 정의
from dataclasses import dataclass
from typing import Optional, List

@dataclass
class NewsArticle:
    title: str                    # 필수 속성
    url: str
    pub_date: Optional[str] = None  # 선택 속성

# 인스턴스 생성
article = NewsArticle("제목", "https://...", "2024-01-20")

# 속성 접근
print(article.title)
```
